```python
!pip install -q xlrd
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as pt
from google.colab import files
import io
uploaded=files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded.get('covid19_Confirmed_dataset.csv')))
df
```

```python
new_df=df.drop(["Lat","Long"],axis=1,inplace=True)
df
```

```python
corona_dataset_aggregated=df.groupby("Country/Region").sum()
corona_dataset_aggregated
```

```python
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Russia"].plot()
pt.legend()
pt.show()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Russia"].diff().max()
```

```python
countries=list(corona_dataset_aggregated.index)
max_infection_rate=[]
for i in countries:
  max_infection_rate.append(corona_dataset_aggregated.loc[i].diff().max())
corona_dataset_aggregated["MAX_INFECTION_RATE"]=max_infection_rate
```

```python
corona_dataset_aggregated
```

```python
corona_data=pd.DataFrame(corona_dataset_aggregated["MAX_INFECTION_RATE"])
```
